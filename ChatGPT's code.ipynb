```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
train_df = train_df.drop(['id', 'keyword', 'location'], axis=1)
train_df['text'] = train_df['text'].str.replace('[^a-zA-Z]', ' ')
```

```python
test_df = test_df.drop(['id', 'keyword', 'location'], axis=1)
test_df['text'] = test_df['text'].str.replace('[^a-zA-Z]', ' ')
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_df['text'], train_df['target'], test_size=0.2, random_state=42)
```

```python
pipeline = Pipeline([
('vect', CountVectorizer(stop_words='english')),
('tfidf', TfidfTransformer()),
('clf', RandomForestClassifier(n_estimators=100, random_state=42)),
])
pipeline.fit(X_train, y_train)
```

```python
print(f"Accuracy score: {pipeline.score(X_test, y_test)}")
```

```python
predictions = pipeline.predict(test_df['text'])
test_df['target'] = predictions
test_df[['id', 'target']].to_csv('solution.csv', index=False)
```
